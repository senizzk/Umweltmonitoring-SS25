```python
import requests
import pandas as pd
from datetime import datetime, timedelta, timezone

senseBoxId = "60a048f7a877b3001b1f9995"
sensorId = "60a048f7a877b3001b1f9996"

end_date = datetime.now(timezone.utc)
start_date = datetime.fromisoformat('2021-05-15T22:19:35.636+00:00')  

all_data = []


current_start = start_date
while current_start < end_date:
    current_end = min(current_start + timedelta(days=7), end_date)
    from_date_str = current_start.strftime("%Y-%m-%dT%H:%M:%S.000Z")
    to_date_str = current_end.strftime("%Y-%m-%dT%H:%M:%S.000Z")
    
    url = f"https://api.opensensemap.org/boxes/{senseBoxId}/data/{sensorId}?from-date={from_date_str}&to-date={to_date_str}&format=json"
    print(f"Çekiliyor: {from_date_str} - {to_date_str}")
    
    try:
        response = requests.get(url, timeout=60)
        response.raise_for_status()
        data = response.json()
        
        df = pd.DataFrame(data)
        if not df.empty:
            df["createdAt"] = pd.to_datetime(df["createdAt"])
            df["hour"] = df["createdAt"].dt.floor("h")
            hourly = df.sort_values("createdAt").groupby("hour").tail(1)
            all_data.append(hourly[["createdAt", "value"]])
    except Exception as e:
        print(f"Hata ({from_date_str}): {e}")
    
    current_start = current_end

# Tüm haftalar birleşiyor
final_df = pd.concat(all_data)
#final_df.to_csv("3_years_hourly_temperature.csv", index=False)

print("Tamamlandı: '3_years_hourly_temperature.csv'")
```
